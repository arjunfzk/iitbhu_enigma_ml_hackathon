```python
def remove_outlier(df_in, col_name, train):
    q1 = train[col_name].quantile(0.25)
    q3 = train[col_name].quantile(0.75)
    iqr = q3-q1
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out

train1=train.copy()
train1=remove_outlier(train1, 'Answers', train)
train1=remove_outlier(train1, 'Views', train)
train1=remove_outlier(train1, 'Reputation', train)
```
